# Labeling dataframe

```python
# Find common file path

csv_path = os.path.join('data', 'csv')
csv_path
```

```python
# Get only .csv from file path 

names = [name for name in os.listdir(csv_path) if name.endswith('.csv')]
names
```

```python
# Remove unnecessary parts of string and put into a list

table_names = []

for n in names:
    table_names.append(n.replace('.csv', '').replace('olist_', '').replace('_dataset', ''))

table_names
```

```python
# Create dictionary from table names and their associated data frames.

data = {}

for k, v in zip(table_names, names):
    data[k] = pd.read_csv(os.path.join(csv_path, v))

data['orders']    
```
